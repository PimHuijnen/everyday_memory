```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
```

```python
paper = 'De Volkskrant'
paper_short = 'volkskrant'
wd = os.getcwd()
path = wd + "/n_years_ago/"
allfiles = glob.glob(os.path.join(path, "*.csv"))
```

```python
for file in allfiles:
    print(file)
```

```python
vk = path + 'volkskrant_n_years_ago_count.csv'
```

```python
df = pd.read_csv(vk, sep='\t', index_col=False)
```

```python
print(df.head())
```

```python
numbers = {1:'een', 
           2:'twee', 
           3:'drie', 
           4:'vier', 
           5:'vijf', 
           6:'zes', 
           7:'zeven', 
           8:'acht', 
           9:'negen', 
           10:'tien', 
           11:'elf', 
           12:'twaalf', 
           13:'dertien', 
           14:'veertien', 
           15:'vijftien', 
           16:'zestien', 
           17:'zeventien', 
           18:'achttien', 
           19:'negentien', 
           20:'twintig', 
           30:'dertig', 
           40:'veertig', 
           50:'vijftig', 
           60:'zestig', 
           70:'zeventig', 
           80:'tachtig', 
           90:'negentig', 
           100:'honderd', 
           110:'honderdtien', 
           120:'honderdtwintig', 
           130:'honderddertig', 
           140:'honderdveertig', 
           150:'honderdvijftig', 
           160:'honderdzestig', 
           170:'honderdzeventig', 
           180:'honderdtachtig', 
           190:'honderdnegentig', 
           200:'tweehonderd'}
```

```python
def word_to_number(getal):
    for number, word in numbers.items():
        if getal == number:
            number_word = word
    return number_word
```

```python
years_ago = [1, 20, 180]
```

```python
plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(15, 10))

for year in years_ago:
    
# Set data
    plt.plot(df.date, df[word_to_number(year) + '_jaar_geleden_permill'], label = '%s jaar geleden' %(word_to_number(year)))

# Set ticks
ax.set_xticks(df.date[df.date % 2 == 0])
plt.xticks(rotation=45)

# Set labels and legend
plt.title("Overall frequency of 'year ago' bigrams in %s (sample size per year = 10,000 documents; per million words)" %(paper), size=14)
plt.xlabel("Year", size=14)
plt.ylabel("Per million words", size=14)
plt.legend()

plt.show()
    
```
